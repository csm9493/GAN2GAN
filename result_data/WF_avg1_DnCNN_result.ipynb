```python
from matplotlib import pyplot as plt
import scipy.io as sio
import numpy as np
```

```python
SMALL_SIZE = 13
MEDIUM_SIZE = 16
BIGGER_SIZE = 24

plt.rc('font', size=MEDIUM_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=BIGGER_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=BIGGER_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=MEDIUM_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=MEDIUM_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=MEDIUM_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title

line_markers = ['-.','-,','-o','-v','-^','-<','->','-1','-2','-3','-4','-s','-p','-*','-h','-H','-+','-x','-D','-d','-|','-_']
```

```python
iter_num = 3
```

```python
tr_loss_arr = []
te_loss_arr = []
psnr_arr = []
ssim_arr = []
denoiser_img_arr = []

for i in range(iter_num+1):
    file_name = 'WF_avg1_DnCNN_iter_'+str(i)+'_result'
    f = sio.loadmat(file_name)
    
    tr_loss_arr.append(f['tr_loss_arr'])
    te_loss_arr.append(f['te_loss_arr'])
    psnr_arr.append(f['psnr_arr'])
    ssim_arr.append(f['ssim_arr'])
    denoiser_img_arr.append(f['denoised_img_arr'])
    
```

```python
plt.figure(figsize=[10,8])

final_ep_psnr_arr = []

for i in range(0, iter_num+1):
    final_ep_psnr_arr.append(psnr_arr[i][0][-1])
        
plt.plot(final_ep_psnr_arr,line_markers[1], linewidth=2.5, label='G2G')
plt.grid(True)
plt.xlabel('Iter')
plt.ylabel('PSNR')
plt.legend(loc='best')
plt.show()
```

```python
plt.figure(figsize=[10,8])

final_ep_ssim_arr = []

for i in range(0, iter_num+1):
    final_ep_ssim_arr.append(ssim_arr[i][0][-1])
        
plt.plot(final_ep_ssim_arr,line_markers[1], linewidth=2.5, label='G2G')
plt.grid(True)
plt.xlabel('Iter')
plt.ylabel('SSIM')
plt.legend(loc='best')
plt.show()

```
