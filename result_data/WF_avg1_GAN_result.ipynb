```python
from matplotlib import pyplot as plt
import scipy.io as sio
import numpy as np
```

```python
SMALL_SIZE = 13
MEDIUM_SIZE = 16
BIGGER_SIZE = 24

plt.rc('font', size=MEDIUM_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=BIGGER_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=BIGGER_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=MEDIUM_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=MEDIUM_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=MEDIUM_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title

line_markers = ['-.','-,','-o','-v','-^','-<','->','-1','-2','-3','-4','-s','-p','-*','-h','-H','-+','-x','-D','-d','-|','-_']
```

```python
file_name = 'WF_avg1_GAN_result'
f = sio.loadmat(file_name)
```

```python
W_c1_arr = f['W_c1_arr']
W_c2_arr = f['W_c2_arr']

plt.figure(figsize=[10,8])

plt.plot(W_c1_arr[0],line_markers[1], linewidth=2.5, label='Critic 1')
plt.plot(W_c2_arr[0],line_markers[1], linewidth=2.5, label='Critic 2')
        
plt.grid(True)
plt.xlabel('Epoch')
plt.ylabel('W_distance')
plt.legend(loc='best')
plt.show()
```

```python
L_g1_arr = f['L_g1_arr']
L_g2_arr = f['L_g2_arr']

plt.figure(figsize=[10,8])

plt.plot(L_g1_arr[0],line_markers[1], linewidth=2.5, label='Generator 1')
plt.plot(L_g2_arr[0],line_markers[1], linewidth=2.5, label='Generator 2')
        
plt.grid(True)
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend(loc='best')
plt.show()
```

```python
L_cycle_arr = f['L_cycle_arr']

plt.figure(figsize=[10,8])

plt.plot(L_cycle_arr[0],line_markers[1], linewidth=2.5, label='Cycle Loss')
        
plt.grid(True)
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend(loc='best')
plt.show()
```

```python
PSNR_arr = f['PSNR_arr']

plt.figure(figsize=[10,8])

plt.plot(PSNR_arr[0],line_markers[1], linewidth=2.5, label='PSNR')
        
plt.grid(True)
plt.xlabel('Epoch')
plt.ylabel('PSNR')
plt.legend(loc='best')
plt.show()
```

```python
mean_n_arr = f['mean_n_arr']
mean_n_hat_arr = f['mean_n_hat_arr']

plt.figure(figsize=[10,8])

plt.plot(mean_n_arr[0],line_markers[1], linewidth=2.5, label='Mean(n)')
plt.plot(mean_n_hat_arr[0],line_markers[1], linewidth=2.5, label='Mean(n_hat)')
        
plt.grid(True)
plt.xlabel('Epoch')
plt.ylabel('Mean')
plt.legend(loc='best')
plt.show()
```

```python
std_n_arr = f['std_n_arr']
std_n_hat_arr = f['std_n_hat_arr']

plt.figure(figsize=[10,8])

plt.plot(std_n_arr[0],line_markers[1], linewidth=2.5, label='Std(n)')
plt.plot(std_n_hat_arr[0],line_markers[1], linewidth=2.5, label='Std(n_hat)')
        
plt.grid(True)
plt.xlabel('Epoch')
plt.ylabel('Std')
plt.legend(loc='best')
plt.show()
```
